```
import pandas as pd

# Import CSV's
subjects_dc = '/home/ad_files/subjects_dc.csv'
adrc_dc = '/home/ad_files/adrc_dc.csv'
freesurfers_dc = '/home/ad_files/freesurfers_dc.csv'

# Read CSV
subjects_df = pd.read_csv(subjects_dc, header=0)
adrc_df = pd.read_csv(adrc_dc, header=0)
freesurfers_df = pd.read_csv(freesurfers_dc, header=0)
```

```
#adrc_df=adrc_df.fillna(0)
adrc_df['height (inches)' ] = adrc_df['height (inches)'].fillna(0)
adrc_df['weight (lbs)' ] = adrc_df['weight (lbs)'].fillna(0)
adrc_df
```

```
grouped_df = adrc_df.groupby("Subject")
adrc_df1= grouped_df["dx1","ADRC_ADRCCLINICALDATA ID"].agg(lambda column: column.str.cat(sep=','))
adrc_df2=grouped_df['ageAtEntry',"mmse",'height (inches)','weight (lbs)'].mean()
adrc_mean=pd.merge(adrc_df1, adrc_df2, on='Subject')
adrc_mean
```

```
freesurfers_df
```

```
grouped_df = freesurfers_df.groupby("Subject")
freesurfers_df1= grouped_df["FS_FSDATA ID","Session"].agg(lambda column: column.str.cat(sep=','))
freesurfers_df2=grouped_df['Left-Hippocampus_volume',"Right-Hippocampus_volume"].mean()
freesurfers_mean=pd.merge(freesurfers_df1, freesurfers_df2, on='Subject')
freesurfers_mean
```

# New Section

```
mean_ad=pd.merge(pd.merge(subjects_df,adrc_mean, on='Subject'),freesurfers_mean,on ='Subject')
```
